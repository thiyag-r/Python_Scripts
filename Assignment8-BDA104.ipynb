```python
import pandas as pd
import numpy as np
```

```python
Dimded = pd.read_csv("C:/Users/santh/Downloads/Data.csv")
```

```python
X = Dimded.iloc[:,1:257]
Y = Dimded.iloc[:,-1]
```

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
rfe = RFECV(estimator=LogisticRegression(),cv=4,scoring='accuracy')
rfe = rfe.fit(X_train,y_train)
cols = X_train.columns[rfe.support_]
acc = accuracy_score(y_test, rfe.estimator_.predict(X_test[cols]))
print('Number of features selected: {}'.format(rfe.n_features_))
print('Test Accuracy {}'.format(acc))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
Fdf = pd.concat([principalDf,Dimded['gnd']], axis = 1)
```

```python
Fdf.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
#plt.xticks(fontsize=12)
#plt.yticks(fontsize=14)
ax.set_xlabel('Principal Component - 1',fontsize=15)
ax.set_ylabel('Principal Component - 2',fontsize=15)
ax.set_title("Principal Component Analysis of Dataset",fontsize=20)
targets = [0,1,2,3]
colors = ['r','g','b','y']
for target, color in zip(targets,colors):
    ind = Fdf['gnd'] == target
    ax.scatter(Fdf.loc[ind, 'principal component 1'], Fdf.loc[ind, 'principal component 2'],c = color, s = 50)
ax.legend(targets)
```

In the graph that has PC1 and PC2, the Class 1 (color coded green) is not that very distant from each other. Hence the variablitly isnt much between them. However in Class 0(color coded red) the points are scattered to some extend showing that there is difference between them. Overall this is the best representation of the samples in the PCA model.

```python
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
tsne = TSNE(n_components=2)
tsne_results = tsne.fit_transform(X)
tsnecomponents = pd.DataFrame(data=tsne_results, columns = ['tsnec1','tsnec2'])
```

```python
tdf = pd.concat([tsnecomponents,Y],axis=1)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(
    x="tsnec1", y="tsnec2",
    hue="gnd",
    palette=sns.color_palette("hls", 4),
    data=tdf,
    legend="full",
    alpha=0.3
)
```

In the TNSE graph, the samples are clearly segmented without overlapping with each other i.e. other different classes. Using this algorithm we can explore the data and have a better insight about the final results of the samples.
