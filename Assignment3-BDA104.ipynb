```python
import json
from pandas.io.json import json_normalize
```

```python
def categorical_genres(df):
    arr = df['genres']
    genre_temp_all = []
    features_genres = ['Action', 'Adventure', 'Fantasy', 'Science Fiction' ]
    for x in arr:
        loaded_json = json.loads(x)
        genres_all_single_movie = []
        for i in range(len(loaded_json)):
            if (loaded_json[i]['name'] in features_genres):
                genres_all_single_movie.append(loaded_json[i]['name'])
            elif (loaded_json[i]['name'] not in features_genres) and (loaded_json[i]['name'] != ''):
                genres_all_single_movie.append("Other")    
        
        genre_temp_all.append(','.join(genres_all_single_movie))
    
    df_genres = pd.DataFrame({'Genres_all':genre_temp_all})   
    return df_genres
```

```python
df_genres = categorical_genres(df)
```

```python
df = pd.concat([df, df_genres['Genres_all'].str.get_dummies(sep=',')], axis=1)
```
