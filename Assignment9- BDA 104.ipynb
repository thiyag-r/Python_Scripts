K-MEAN

K-Mean is one of the category of centroid based clustering. Centroid is nothing but a data point at the center of a cluster.
The k-means algorithm searches for a pre-determined number of clusters within an unlabeled multidimensional dataset.

```python
import numpy as np
import pandas as pd
import random
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
```

```python
for i in range(100):
    center_value = np.random.randint(2,10)
    clusterstd = np.random.uniform(0.01,0.99)
    X, y = make_blobs(n_samples=300, centers=center_value,random_state=0, cluster_std=clusterstd)
    plt.scatter(X[:, 0], X[:, 1], s=50);
    SS = {}
    for k in range(2,15):
        k_model = KMeans(n_clusters=k, random_state=0)
        pred = k_model.fit_predict(X)
        Score = silhouette_score(X, pred, metric='euclidean', sample_size=None)
        SS .update({k:Score})
        print("Centers:", center_value, "kValue:", max(SS, key = SS.get), "silhouette_score:" ,max(SS.values()))
```
